```python
import random

# -------------------------
# CONFIGURAÇÕES
# -------------------------

population_size = 200
generations = 30
mutation_rate = 0.1  # 10% chance de mutação por característica

# -------------------------
# CRIAR POPULAÇÃO INICIAL
# -------------------------

def create_random_neebler():
    return {
        "size": random.choice(["small", "big"]),
        "speed": random.choice(["slow", "fast"]),
        "resistance": random.choice(["low", "high"])
    }

population = [create_random_neebler() for _ in range(population_size)]

# -------------------------
# FUNÇÃO DE SOBREVIVÊNCIA
# -------------------------

def survival_chance(neebler):

    chance = 0.5  # base

    # tamanho
    if neebler["size"] == "small":
        chance += 0.1
    else:
        chance -= 0.05

    # velocidade
    if neebler["speed"] == "fast":
        chance += 0.1

    # resistência
    if neebler["resistance"] == "high":
        chance += 0.15

    # fator ambiental aleatório
    chance += random.uniform(-0.1, 0.1)

    return max(0, min(1, chance))

# -------------------------
# FUNÇÃO DE MUTAÇÃO
# -------------------------

def mutate_trait(trait, options):
    if random.random() < mutation_rate:
        return random.choice(options)
    return trait

# -------------------------
# SIMULAÇÃO
# -------------------------

for generation in range(generations):

    survivors = []

    # --- SOBREVIVÊNCIA ---
    for neebler in population:
        if random.random() < survival_chance(neebler):
            survivors.append(neebler)

    if len(survivors) == 0:
        print(f"Geração {generation + 1}: Todos morreram.")
        break

    # --- REPRODUÇÃO ---
    new_population = []

    for parent in survivors:
        num_babies = random.randint(0, 3)

        for _ in range(num_babies):
            baby = {
                "size": mutate_trait(parent["size"], ["small", "big"]),
                "speed": mutate_trait(parent["speed"], ["slow", "fast"]),
                "resistance": mutate_trait(parent["resistance"], ["low", "high"])
            }
            new_population.append(baby)

    population = new_population

    # Estatísticas
    small = sum(1 for n in population if n["size"] == "small")
    big = sum(1 for n in population if n["size"] == "big")
    fast = sum(1 for n in population if n["speed"] == "fast")
    high_res = sum(1 for n in population if n["resistance"] == "high")

    print(f"Geração {generation + 1}")
    print("População:", len(population))
    print("Small:", small, "| Big:", big)
    print("Fast:", fast)
    print("High Resistance:", high_res)
    print("-" * 40)

    if len(population) == 0:
        print("Extinção total.")
        break
```
